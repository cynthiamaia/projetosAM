## 1. Realizar um estudo sobre as funções de ativações informando qual a aplicabilidade ou indicação de utilização de cada uma.



As redes neurais são um algoritmo de aprendizado baseado em conexões existentes entre nós, formando uma espécie de rede, similar aos neurônios do cérebro humano, as quais permitem um aprendizado de padrões, correlações, agrupamento e classificação de dados através dos pesos utilizados nos nós.

O desempenho da rede neural depende de alguns fatores como: número de camadas ocultas, algoritmo de aprendizado e a função de ativação.  A escolha da função de ativação pode influenciar o desempenho da rede neural e sua complexidade [1]. As funções de ativação tem como objetivo de limitar a amplitude válida do sinal de saída do neurônio a um valor finito [6] e permitem soluções de problemas mais complexos, realizando uma transformação não-linear dos dados de entrada para que sejam capazes de aprender e executar tarefas mais complexas. A saída transformada de cada neurônio é enviada ao neurônio presente na próxima camada de neurônios como entrada, conferindo assim essa capacidade não-linear no processamento pela RNA. 

As funções de ativação são importantes diante de que, para muitas tarefas complexas, transformações lineares não seriam capazes de executar um aprendizado satisfatório[2]. Existem diferentes tipos de funções de ativação, como: Rectified Linear Unit (ReLU),  Leaky Rectified Linear Unit (Leaky ReLU),  ELU (Exponential Linear Unit), Sigmoide, Tangente Hiperbólica (tanh), Logística, Linear, Gaussiana e Maxout. 

**ReLU** é uma função de ativação não-linear amplamente utilizada por sua simplicidade e seu relativo baixo custo computacional, já que na ReLU os neurônios simplesmente são ativados quando o valor é positivo e não ativados quando o valor é negativo. No entanto, esta função sofre de um problema que é conhecido como: Dying ReLUs, em que, ao longo do treinamento, muitos neurônios morrem, ou seja, passam a ter saída sempre nula. Isso acontece especialmente se a taxa de aprendizado for grande, em que o gradiente é zero para z < 0, o que faz com que os neurônios morram, mas, sua capacidade de gerar zero pode ser útil em alguns casos [3].

Já como uma solução do problema supracitado de Dying ReLUs, propõe-se a função **Leaky ReLU**, que é uma variante da ReLU, em que o hiperparâmetro $\alpha$ (taxa de aprendizado) define o quanto a função “vaza” quando $ z<0 $. (em outras palavras, $\alpha$ é a inclinação da função), o que permite que os neurônios nunca morram, sendo configurada tipicamente para 0.01 ou valor próximo, removendo-se o gradiente zero. Se é necessário que a rede seja a mais rápida possível, o uso da Leaky ReLU é indicado [2].

Outro tipo de função de ativação é a **ELU**, que superou as versões ReLU, sendo bastante semelhante a ReLU, mas com algumas diferenças, como: evita o problema das unidades mortas, já que tem um gradiente diferente de zero para z<0. Assume valores negativos que permitem aproximar a ativação da unidade para mais próximo de zero e com isso é reduzida a complexidade computacional e é melhorada a velocidade do aprendizado [4]. Uma desvantagem é que devido ao uso da função exponencial, o seu cálculo é mais lento do que o da ReLU e suas variantes quando o modelo já treinado é aplicado a novas instâncias, entretanto, no treinamento tem-se uma convergência mais rápida do que a ReLU. Sendo assim, se o fator de desempenho em tempo de execução importa, é preferível as Leaky ReLUs do que as ELUs [3].

A função de ativação **Sigmoide**, tem a vantagem de apresentar uma curva suave e continuamente diferenciável e isso facilita o processo de treinamento, no qual, possibilita direcionamento  para adequação dos pesos de acordo com erro do sistema. Bastante utilizada em camadas que precisam ter em suas saídas números positivos, diante que o valor de saída é entre 0 e 1, funcionam bem ao lidar com problemas de classificação binária [5].

A **tangente hiperbólica** é semelhante a Sigmóide, com a diferença que os valores de saída ficam entre -1 e 1. Se é necessário gerar um número entre -1 e 1 a tangente pode ser útil na camada de saída, mas não são muito utilizadas para as camadas ocultas. 

A função de ativação **logística** é útil quando se precisa estimar uma probabilidade na camada de saída, sendo raramente usada nas camadas ocultas [3]. Já a função Linear, em que não fornece não linearidade e é ideal para tarefas simples, em que deseja-se ter uma alta interpretação, não conseguindo lidar com a complexidade dos dados, sendo útil para tarefas simples de aprendizado [7]. 

Já acerca do uso da função de ativação **gaussiana**, a saída do neurônio fornece resultados iguais para valores posicionados a uma mesma distância de seu centro (média), sendo que a curva é simétrica em relação a este [8]. O parâmetro c da função delimita o centro e já o parâmetro σ é em relação ao desvio da função, que corresponde o quanto a curva está dispersa em relação ao centro. Com valores pequenos de σ, produzirá uma gaussiana mais estreita, assim, com uma distribuição mais próxima do centro. Já com valores grandes de σ irá produzir uma maior espalhamento [9].

Já a função de ativação **Maxout** funciona bem na prática, em que sua saída é o máximo (max) das entradas [10]. É uma generalização da RELU, com benefícios da unidade RELU (sem saturação, linearidade) e não tem a desvantagem dos neurônios morrendo, mas, tem a desvantagem de o que número de parâmetros dobra para cada neurônio levando a um alto número total de parâmetros [11]. Sendo adequado para treinamento com evasão e tem uma capacidade de ajuste que é forte que é capaz  de se ajustar a qualquer função convexa [12].

### Pré-Processamento

As técnicas de pré-processamento tem por objetivo modificar os dados de entrada para um formato que melhore a qualidade dos dados e, consequentemente, possibilite realizar análises com um melhor desempenho. Foi verificado o desbalanceamento das classes, constatando-se que o número de instâncias por classe varia bastante [9,76]. A partir disso, foi aplicado o SMOTE para realizar o balanceamento das classes, em que se criou instâncias sintéticas das classes minoritárias. Cada uma delas ficou com 76 instâncias. Outro fator observado, foi que os dados se apresentaram em escalas bem diferentes. No entanto, para esse modelo, não foi realizada a etapa de normalização de dados. Isso porque as funções de ativação propiciam que as saídas do modelo já se tornem na escala [0,1].


````
# Contagem de instâncias por classe:
Counter(df['class'])
````

````
# Histograma de frequencia de instâncias por classe:
# definindo os eixos 
x_list=[1, 2, 3, 5, 6, 7]
y_list=[70, 76, 17, 13, 9, 29]
# plotagem do grafico
plt.rcParams['xtick.labelsize'] = 22
plt.rcParams['ytick.labelsize'] = 22
plt.figure(figsize=(15,10))
plt.bar(x_list,y_list, color='green')
# informaçoes
plt.ylabel("Frequência")
plt.xlabel("Classes")
plt.rcParams.update({'font.size': 22})
plt.title("Gráfico de Frequência das classes")
````

````
# Separando X e y:
X = df[df.columns[:-1]].to_numpy()
y = df[df.columns[[-1]]].to_numpy().ravel()
````

````
# Vamos realizar o balanceamento das classes por meio de sobreamostragem das classes
# até que fiquem todas com 76 instâncias cada. Já que 76 é o número de instâncias da classe majoritária:
X, y = SMOTE().fit_resample(X, y)
Counter(y)
````

````
# Separando os 5 folds para validação cruzada repetida 5 vezes:
#rkf = RepeatedKFold(n_splits=5, n_repeats=5, random_state=0)
# Separando os 5 folds para validação cruzada
skf = StratifiedKFold(n_splits=5, shuffle=False)
X_train_folds = []
y_train_folds = []
X_test_folds = []
y_test_folds = []
for train_index, test_index in skf.split(X,y):
  X_train_folds.append(X[train_index])
  X_test_folds.append(X[test_index])
  y_train_folds.append(y[train_index])
  y_test_folds.append(y[test_index])


X_train, val_x, y_train, val_y = train_test_split(X_train_folds, y_train_folds, test_size=0.20)

len(X_train_folds) 
````

````
def validacao_cruzada(model):
  cv_results = {'test_accuracy':[]}
  for fold in range(len(X_train_folds)):
      X_train = X_train_folds[fold]
      y_train = y_train_folds[fold]
      X_test = X_test_folds[fold]
      y_test = y_test_folds[fold]
      #print('Contagem de classes:\nTreino = ', Counter(y_train), '\nTeste: ', Counter(y_test))

      model.fit(X_train, y_train)
      y_pred = model.predict(X_test)

      test_acc = accuracy_score(y_test, y_pred)
      cv_results['test_accuracy'].append(test_acc)
  return cv_results
````

### Construção dos modelos

Temos 9 camadas atributos de entrada e uma única variável alvo (class) com seis possíveis valores, então vamos criar modelos com redes contendo 9 nós na camada de entrada e seis nós na camada de saída. A camada de saída, para suportar multiclasse, será implementada utilizando a função softmax.

Para nosso modelo baseline, teremos uma camadas escondida (hidden layer) contendo apenas 4 nós. A arquitetura deste modelo está representada na Tabela e Figura abaixo.

| Layer (type)     | Output Shape | Param # |
|------------------|--------------|---------|
| dense_21 (Dense) | (None, 4)    | 40      |
| dense_22 (Dense) | (None, 6)    | 30      |

Executaremos este modelo com apenas 25 épocas.



modelo_baseline.png


````
def baseline_model():
  model = Sequential() 
  #Primeira camada escondida:
  model.add(Dense(4, input_dim=9, activation='relu'))
  #Camada de saída:
  model.add(Dense(6, activation='softmax'))
  #Compilando a rede:
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
  #print(model.summary())
  return model
````

Agora vamos construir modelos mais complexos alterando o número de camadas escondidas para 2 e variando outros parâmetros para cada modelo. Construiremos diversos modelos, todos com a mesma arquitetura (conforme tabela e figura abaixo), em que mudaremos apenas a função de ativação das camadas intermediárias.

| Layer (type)     | Output Shape | Param # |
|------------------|--------------|---------|
| dense_23 (Dense) | (None, 6)    | 60      |
| dense_24 (Dense) | (None, 6)    | 42      |
| dense_24 (Dense) | (None, 6)    | 42      |

Total params: 144

Trainable params: 144

Non-trainable params: 0

Os modelos construídos com essa arquitetura serão executados com 500 épocas.


modelos_complexos.png

````
# Função de ativação Relu:
def complex_model_1():
  model = Sequential() 
  #Primeira camada escondida:
  model.add(Dense(6, input_dim=9, activation='relu'))
  #Segunda camada escondida:
  model.add(Dense(6, activation='relu'))
  #Camada de saída
  model.add(Dense(6, activation='softmax'))
  #Compilando a rede:
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
  #print(model.summary())
  return model
````

````
# Função de ativação  Tangente Hiperbólica : 
def complex_model_2():
  model = Sequential() 
  #Primeira camada escondida:
  model.add(Dense(6, input_dim=9, activation='tanh'))
  #Segunda camada escondida:
  model.add(Dense(6, activation='tanh'))
  #Camada de saída
  model.add(Dense(6, activation='softmax'))
  #Compilando a rede:
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
  #print(model.summary())
  return model
````

````
# Função de ativação Sigmóide: 
def complex_model_3():
  model = Sequential() 
  #Primeira camada escondida:
  model.add(Dense(6, input_dim=9, activation='sigmoid'))
  #Segunda camada escondida:
  model.add(Dense(6, activation='sigmoid'))
  #Camada de saída
  model.add(Dense(6, activation='softmax'))
  #Compilando a rede:
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
  #print(model.summary())
  return model
````

````
# Função de ativação Elu:
def complex_model_4():
  model = Sequential() 
  #Primeira camada escondida:
  model.add(Dense(6, input_dim=9, activation='elu'))
  #Segunda camada escondida:
  model.add(Dense(6, activation='elu'))
  #Camada de saída
  model.add(Dense(6, activation='softmax'))
  #Compilando a rede:
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
  #print(model.summary())
  return model
````

````
# Função de ativação Leaky Relu:
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

def complex_model_5():
  model = tf.keras.models.Sequential()
  #Primeira camada escondida:
  model.add(layers.Dense(6, input_dim=9, activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
  #Segunda camada escondida:
  model.add(layers.Dense(6, activation=tf.keras.layers.LeakyReLU(alpha=0.2)))
  #Camada de saída:
  model.add(layers.Dense(6, activation='softmax'))
  #Compilando a rede:
  model.compile(optimizer='Adam',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
  return model
````

````
# Função de ativação Linear:
def complex_model_6():
  model = Sequential() 
  #Primeira camada escondida:
  model.add(Dense(6, input_dim=9, activation='linear'))
  #Segunda camada escondida:
  model.add(Dense(6, activation='linear'))
  #Camada de saída
  model.add(Dense(6, activation='softmax'))
  #Compilando a rede:
  model.compile(loss='categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
  #print(model.summary())
  return model
````

````
#Criando dicionário para armazenar resultados:
accuracies = {'baseline':[],'linear':[], 'leakyRelu':[], 'relu':[], 'elu':[], 'tanh':[], 'sigmoid':[]} 
````

````
baseline = KerasClassifier(build_fn=baseline_model, epochs=25, batch_size=5, verbose=0)
cv_results = validacao_cruzada(baseline)
#print('Acurácia média para o modelo baseline:')
#print(np.mean(cv_results['test_accuracy']))
accuracies['baseline'] = cv_results['test_accuracy']
````

````
complex_1 = KerasClassifier(build_fn=complex_model_1, epochs=500, batch_size=5, verbose=0)
cv_results_1 = validacao_cruzada(complex_1)
#print('Acurácia média para o modelo complex_model_1: ReLU')
#print(np.mean(cv_results_1['test_accuracy']))
accuracies['relu'] = cv_results_1['test_accuracy']
````

````
complex_2 = KerasClassifier(build_fn=complex_model_2, epochs=500, batch_size=5, verbose=0)
cv_results_2 = validacao_cruzada(complex_2)
#print('Acurácia média para o modelo complex_model_2: Tanh')
#print(np.mean(cv_results_2['test_accuracy']))
accuracies['tanh'] = cv_results_2['test_accuracy']
````

````
complex_3 = KerasClassifier(build_fn=complex_model_3, epochs=500, batch_size=5, verbose=0)
cv_results_3 = validacao_cruzada(complex_3)
#print('Acurácia média para o modelo complex_model_3: Sigmoid')
#print(np.mean(cv_results_3['test_accuracy']))
accuracies['sigmoid'] = cv_results_3['test_accuracy']
````

````
complex_4 = KerasClassifier(build_fn=complex_model_4, epochs=500, batch_size=5, verbose=0)
cv_results_4 = validacao_cruzada(complex_4)
#print('Acurácia média para o modelo complex_model_4: ELU')
#print(np.mean(cv_results_4['test_accuracy']))
accuracies['elu'] = cv_results_4['test_accuracy']
````

````
complex_5 = KerasClassifier(build_fn=complex_model_5, epochs=500, batch_size=5, verbose=0)
cv_results_5 = validacao_cruzada(complex_5)
#print('Acurácia média para o modelo complex_model_5: LeakyReLU')
#print(np.mean(cv_results_5['test_accuracy']))
accuracies['leakyRelu'] = cv_results_5['test_accuracy']
````

````
complex_6 = KerasClassifier(build_fn=complex_model_6, epochs=500, batch_size=5, verbose=0)
cv_results_6 = validacao_cruzada(complex_6)
#print('Acurácia média para o modelo complex_model_6: Linear')
#print(np.mean(cv_results_6['test_accuracy']))
accuracies['linear'] = cv_results_6['test_accuracy']
````
